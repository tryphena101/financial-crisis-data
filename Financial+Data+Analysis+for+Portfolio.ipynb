```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Deutsche Bank 
DB = data.DataReader("DB", 'google', start, end)

# CitiGroup
C = data.DataReader("C", 'google', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'google', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'google', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'google', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'google', start, end)
```

```python
tickers = ['DB', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([DB, C, GS, JPM, MS, WFC],axis=1,keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close', level=1, axis=1)
```

```python
returns = pd.DataFrame()
```

```python
for x in tickers:
    returns[x+ ' Returns'] = bank_stocks[x]['Close'].pct_change()

returns.dropna(inplace=True)
```
